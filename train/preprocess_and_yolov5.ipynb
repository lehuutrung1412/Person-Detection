## Process Dataset

```python
import json
import os
import shutil
```

```python
train_anno_file = open('/content/drive/MyDrive/Study Documents/CS231 - CV/Project/annotation/instances_train2017.json','r')
train_data = json.load(train_anno_file)
train_anno_file.close()
```

```python
json.load(open('/content/drive/MyDrive/Study Documents/CS231 - CV/Project/annotation/instances_train2017.json','r'))
```

```python
print(train_data.keys())
```

```python
print(len(train_data['annotations'][0]))
```

```python
train_data['annotations'].sort(key= lambda x: x['category_id'])
train_data['images'].sort(key= lambda x: x['id'])
```

```python
train_data_filtered = {}
train_data_filtered['annotations'] = []
for anno in train_data['annotations']:
    if anno['category_id'] == 1:
        left = 0
        right = len(train_data['images']) - 1
        while left <= right:
            mid = (left + right) >> 1
            if train_data['images'][mid]['id'] < anno['image_id']:
                left = mid + 1
            elif train_data['images'][mid]['id'] > anno['image_id']:
                right = mid - 1
            else:
                break
        anno['file_name'] = train_data['images'][mid]['file_name']
        anno['width'] = train_data['images'][mid]['width']
        anno['height'] = train_data['images'][mid]['height']
    else:
        break
    train_data_filtered['annotations'].append(anno)
```

```python
for anno in train_data_filtered['annotations']:
    anno.pop('segmentation')
    anno.pop('iscrowd')
    anno.pop('id')
    anno.pop('category_id')
```

```python
print(len(train_data_filtered['annotations']))
```

```python
print(train_data_filtered['annotations'][10200]['file_name'].split('.')[0])
```

```python
%cd /content/drive/MyDrive/Study Documents/CS231 - CV/Project/labels/
for anno in train_data_filtered['annotations']:
    width_normalized = 1 / anno['width']
    height_normalized = 1 / anno['height']
    x = anno['bbox'][0] + anno['bbox'][2] / 2
    y = anno['bbox'][1] + anno['bbox'][3] / 2
    x *= width_normalized
    y *= height_normalized
    width = anno['bbox'][2] * width_normalized
    height = anno['bbox'][3] * height_normalized
    with open(anno['file_name'].split('.')[0] + '.txt','a') as f:
      res = ' '.join(str(i) for i in [0, x, y, width, height])
      res += '\n'
      f.write(res)
```
